```python
import pandas as pd
import numpy as np
```

```python
df=pd.read_csv('view1.csv')
```

```python
df.head()
```

```python
df.duplicated().sum()
```

```python
df.isnull().sum()
```

```python
df['Value_Deal'].fillna('Unknown', inplace=True)
df['Internet_Type'].fillna('Unknown', inplace=True)
```

```python
df.isnull().sum()
```

```python
df=df.drop(['Customer_ID','Churn_Category','Churn_Reason'],axis=1)
```

```python
from sklearn.preprocessing import LabelEncoder
columns_to_encode = [
    'Gender', 'Married', 'State', 'Value_Deal', 'Phone_Service', 'Multiple_Lines',
    'Internet_Service', 'Internet_Type', 'Online_Security', 'Online_Backup',
    'Device_Protection_Plan', 'Premium_Support', 'Streaming_TV', 'Streaming_Movies',
    'Streaming_Music', 'Unlimited_Data', 'Contract', 'Paperless_Billing',
    'Payment_Method']
label_encoders = {}
for column in columns_to_encode:
    label_encoders[column] = LabelEncoder()
    df[column] = label_encoders[column].fit_transform(df[column])
```

```python
df['Customer_Status']=df['Customer_Status'].map({'Stayed':0,'Churned':1})
```

```python
X=df.drop('Customer_Status',axis=1)
y=df['Customer_Status']
```

```python
from imblearn.over_sampling import SMOTE
```

```python
X_res,y_res=SMOTE().fit_resample(X,y)
```

```python
y_res.value_counts()
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X_res, y_res, test_size=0.25, random_state=42)
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
sc=StandardScaler()
```

```python
X_train=sc.fit_transform(X_train)
X_test=sc.transform(X_test)
```

```python
 X_train
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
log=LogisticRegression()
```

```python
log.fit(X_train,y_train)
```

```python
y_pred1=log.predict(X_test)
```

```python
from sklearn.metrics import accuracy_score
```

```python
from sklearn.metrics import precision_score,recall_score,f1_score
```

```python
print(accuracy_score(y_test,y_pred1))
print(precision_score(y_test,y_pred1))
print(recall_score(y_test,y_pred1))
print(f1_score(y_test,y_pred1))
```

```python
from sklearn import svm
```

```python
svm=svm.SVC()
```

```python
svm.fit(X_train,y_train)
```

```python
y_pred2=svm.predict(X_test)
```

```python
print(accuracy_score(y_test,y_pred2))
print(precision_score(y_test,y_pred2))
print(recall_score(y_test,y_pred2))
print(f1_score(y_test,y_pred2))
```

```python
from sklearn.neighbors import KNeighborsClassifier
```

```python
knn=KNeighborsClassifier()
```

```python
knn.fit(X_train,y_train)
```

```python
y_pred3=knn.predict(X_test)
```

```python
print(accuracy_score(y_test,y_pred3))
print(precision_score(y_test,y_pred3))
print(recall_score(y_test,y_pred3))
print(f1_score(y_test,y_pred3))
```

```python
from sklearn.tree import DecisionTreeClassifier
```

```python
dt=DecisionTreeClassifier()
```

```python
dt.fit(X_train,y_train)
```

```python
y_pred4=dt.predict(X_test)
```

```python
print(accuracy_score(y_test,y_pred4))
print(precision_score(y_test,y_pred4))
print(recall_score(y_test,y_pred4))
print(f1_score(y_test,y_pred4))
```

```python
from sklearn.ensemble import RandomForestClassifier
```

```python
rf=RandomForestClassifier()
```

```python
rf.fit(X_train,y_train)
```

```python
y_pred5=rf.predict(X_test)
```

```python
print(accuracy_score(y_test,y_pred5))
print(precision_score(y_test,y_pred5))
print(recall_score(y_test,y_pred5))
print(f1_score(y_test,y_pred5))
```

```python
xgb=XGBClassifier()
```

```python
xgb.fit(X_train,y_train)
```

```python
y_pred6=xgb.predict(X_test)
```

```python
print(accuracy_score(y_test,y_pred6))
print(precision_score(y_test,y_pred6))
print(recall_score(y_test,y_pred6))
print(f1_score(y_test,y_pred6))
```
